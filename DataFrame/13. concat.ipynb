```python
import numpy as np
import pandas as pd
```

### concat
- default: 행으로 연결
- axis=1: 컬럼으로 연결
- join
    - 내부 조인(inner): 공통된 데이터만 연결
    - 외부 조인(outer): default 값, 모든 데이터 연결

```python
employee = {'empno': [1, 2, 3, 4, 5], 'ename': ['kim', 'lee', 'park', 'song', 'min'], 'dept': [1, 2, 1, 3, 2]}
dept = {'dept': [1, 2, 3], 'deptname': ['관리직', '영업직', '개발직']}
info = {'empno': [1, 3, 2, 5, 4], 'addr': ['서울시', '부산시', '광주시', '광주시', '광주시',],
       'phone': ['010-1111-1111', '010-2222-2222', '010-3333-3333', '010-44444-4444', '010-5555-5555']}
```

```python
emp = pd.DataFrame(employee)
dept = pd.DataFrame(dept)
info = pd.DataFrame(info)
```

```python
# emp, dept 행으로 연결
pd.concat([emp, dept])
```

```python
# emp, dept 컬럼으로 연결
pd.concat([emp, dept], axis=1)
```

```python
# emp, info 행으로 연결
pd.concat([emp, info])
```

```python
# emp, info 컬럼으로 연결
pd.concat([emp, info], axis=1)
```

```python
# emp, dept, info 행으로 연결
pd.concat([emp, dept, info])
```

```python
# emp, dept, info 컬럼으로 연결
pd.concat([emp, dept, info], axis=1)
```

```python
# emp, dept, info 행으로 연결후 인덱스 0부터 다시 지정
pd.concat([emp, dept, info], ignore_index=True)
```

```python
# emp, dept, info 컬럼으로 연결후 컬럼명 0부터 다시 지정
pd.concat([emp, dept, info], axis=1, ignore_index=True)
```

```python
# 공통 컬럼만 연결
pd.concat([emp, dept],  join='inner')
```

```python
# 공통 인덱스만 연결
pd.concat([emp, dept], axis=1, join='inner')
```
