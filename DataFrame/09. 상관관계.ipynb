```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "AppleGothic"
matplotlib.rcParams['axes.unicode_minus'] = False
# matplotlib.rcParams['text.color'] = 'white'
# matplotlib.rcParams['axes.labelcolor'] = 'white'
# matplotlib.rcParams['xtick.color'] = 'white'
# matplotlib.rcParams['ytick.color'] = 'white'
import seaborn as sns
```

### 상관관계
- 2개 변수가 선형 관계가 있는 범위를 표현하는 통계적 측도, 원인과 결과에 관한 표현 없이 간단한 관계를 설명하는 일반적인 도구
- 데이터 간 단순 관계를 설명하는 데 유용

```python
# csv파일 불러오기
# enconding='cp949': 한글 인코딩, 'euc-kr'도 가능
# index_col='관서명': 인덱스를 '관서명'컬럼으로 지정
# thousands=',': 숫자 데이터의 데이터 타입을 문자형이 아닌 숫자형으로 맞춰준다.
df = pd.read_csv('../data1/crime_in_Seoul.csv',
                encoding='cp949', index_col='관서명', thousands=',')
df.head()
```

```python
# 상관관계
df.corr()
```

 -1.0 ~ -0.7: 강한 음적 선형관계
 
 -0.7 ~ -0.3: 뚜렷한 음적 선형관계
 
 -0.3 ~ -0.1: 약한 음적 선형관계
 
 -0.1 ~ 0.1: 거의 무시될 수 있는 선형관계
 
 0.1 ~ 0.3: 약한 양적 선형관계
 
 0.3 ~ 0.7: 뚜렷한 양적 선형관계
 
 0.7 ~ 1.0: 강한 양적 선형관계
 

 양적 선형관계: 비례관계
 
 음적 선형관계: 반비례관계

```python
plt.figure(figsize=(10, 7))
sns.heatmap(df.corr(), cmap='Reds', annot=True, vmin=-1, vmax=1, linewidth=1)
plt.show()
```

```python
# 살인 발생과 폭력 발생 scatter로 그리기 
plt.figure(figsize=(10, 7))
plt.scatter(df['살인 발생'], df['폭력 발생'], c='r', s=50)
plt.xlabel('살인 발생')
plt.ylabel('폭력 발생')
plt.show()
```
