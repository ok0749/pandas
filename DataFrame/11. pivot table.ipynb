```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "AppleGothic"
matplotlib.rcParams['axes.unicode_minus'] = False
import seaborn as sns
```

```python
df = pd.read_csv('../data4/grade.csv')
df
```

```python
# 학년별 평균
df.groupby('학년').mean()
```

```python
# default: mean
# 학년별 평균
df.pivot_table(index='학년')[['중간', '기말']]
```

```python
# 학년별 합계
df.pivot_table(index='학년', aggfunc='count')[['중간', '기말']]
```

```python
# 학년, 과목별 평균
df.groupby(['학년', '과목']).mean()
```

```python
# 학년, 과목별 평균
df.pivot_table(index=['학년', '과목'])[['중간', '기말']]
```

```python
# 학년, 과목별 합계와 평균
df.pivot_table(index=['학년', '과목'], aggfunc=['sum', 'mean'])
```

```python
# 학년, 과목별 함수 적용
# 학년, 과목별 평균이 70보다 크면 합격, 70보다 작거나 같으면 불합격 출력
df.pivot_table(index=['학년', '과목'], aggfunc=lambda x: '합격' if x.mean() > 70 else '불합격')[['중간', '기말']]
```

```python
# 컬럼을 '중간'컬럼으로 지정
# 해당되지 않는 값은 NaN으로 출력
df.pivot_table(index=['학년', '과목'], columns='중간')
```

```python
# 컬럼을 '결과'컬럼으로 지정
df.pivot_table(index=['학년', '과목'], columns='결과')
```
