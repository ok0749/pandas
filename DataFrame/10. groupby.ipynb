```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "AppleGothic"
matplotlib.rcParams['axes.unicode_minus'] = False
import seaborn as sns
```

```python
# excel파일 불러오기
df = pd.read_excel('../data1/hotel.xlsx', index_col=0)
df.head()
```

```python
# 데이터 프레임의 값들의 개수 확인(결측치 제외)
df.value_counts()
```

```python
# 각 grade에 해당하는 인덱스
df.groupby('grade').groups
```

```python
# grade별 합계
df.groupby('grade').sum()
```

```python
# grade별 최대값
df.groupby('grade').max()
```

```python
# price별 개수
# 결측치는 count하지 않는다
df.groupby('price').count()
```

```python
# 두 개이상의 통계 함수를 쓸때는 agg 함수 사용한다.
df.groupby('grade').agg(['sum', 'mean']).round(2)
```

```python
df2 = df.groupby('grade')['price'].agg(['sum', 'mean'])
df2.columns = ['가격의 총합', '가격의 평균']
df2
```
