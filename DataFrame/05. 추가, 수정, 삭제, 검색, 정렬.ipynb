```python
import numpy as np
import pandas as pd
```

### 추가, 수정

```python
data = {'eng': [10, 30, 50, 70],
       'kor': [20, 40, 60, 80],
       'math': [90, 50, 20, 70]}
df = pd.DataFrame(data, index=['a', 'b', 'c', 'd'])
df
```

```python
# 컬럼 추가 
# 컬럼 있으면 수정, 없으면 추가
df['my'] = [1, 2, 3, 4]
df
```

```python
# 행 추가
# 행 있으면 수정, 없으면 추가
df.loc['e'] = [80, 90, 100, 5]
df
```

### 삭제

```python
# 행 삭제
# 결과 반영하려면 변수로 받거나 inplace=True
df.drop(index=['a', 'c'])
df.drop(['a', 'c'])
```

```python
# 컬럼 삭제
df.drop(columns=['eng', 'math'])
df.drop(['eng', 'math'], axis=1)
```

```python
# 행, 컬럼 삭제
df.drop(index=['a', 'c'], columns=['eng', 'math'])
```

### 정렬

```python
# 인덱스 기준 오름차순 정렬
# 결과 반영하려면 변수로 받거나 inplace=True
df.sort_index()
```

```python
# 인덱스 기준 내림차순 정렬
df.sort_index(ascending=False)
```

```python
# 특정 컬럼 기준 오름차순 정렬
df.sort_values(by='math')
```

```python
# 특정 컬럼 기준 내림차순 정렬
df.sort_values(by='math', ascending=False)
```

```python
# 크기가 같은 데이터의 경우 다음 컬럼의 데이터 값을 참고한다.
df['math'] = 20, 50, 70, 70, 100
df.sort_values(by=['math', 'kor'], ascending=False)
```
